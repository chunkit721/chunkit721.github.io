```python
import pandas as pd
import numpy as np
db = pd.read_csv('hsi_put.csv')

print db['Warrant Ticker'][0]
```

```python
from lxml import html,cssselect
import requests
import sys

stocks = db['Warrant Ticker']
#stocks = ['26554']

def get_quote(name,option1):
    page = requests.get('http://warrants.ubs.com/en/warrants/warrants_analyzer_performance/code/' + name)
    tree = html.fromstring(page.content)
    ex_price_pattern = '//*[@id="warrants_analyzer_performance"]/div[5]/table/tbody/tr[5]/td/text()'
    ticker_names_pattern = '//*[@id="warrants_analyzer_performance"]/h1/text()'
    
    ex_price = tree.xpath(ex_price_pattern)
    ticker_names = tree.xpath(ticker_names_pattern)
    string = ""
    
    if option1 == "price":
        for foo in ex_price:
            return foo
    elif option1 == "name":
        for foo in ticker_names:
            return foo
    else:
        return "No result"
    
name_list = list()
price_list = list()
for name in stocks:
    name_list.append(get_quote(str(name),"name"))
    price_list.append(get_quote(str(name),"price"))

db['Name'] = name_list
db['Exercise Price'] = price_list

db
```

```python
hsi_list = list()
for name in db['Name']:
    if "HSI" in name:
        hsi_list.append(True)
    else:
        hsi_list.append(False)
db['hsi'] = hsi_list
db
```
